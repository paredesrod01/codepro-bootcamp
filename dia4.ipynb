```python
class Perro:

    especie= 'perro'

    def __init__(self, nombre, color, caracteristica):
        self.nombre= nombre
        self.color= color
        self.caracteristica= caracteristica

#crear el objeto
firulais= Perro('firulais', 'marron', 'ladra')
print(f'cuando yo hablo Maximo el {firulais.especie} {firulais.color} {firulais.caracteristica}')

momo= Perro('Momo', 'negro', 'mueve la cola')
print(f'cuando yo digo PASEO! Momo el {momo.especie} {momo.color} {momo.caracteristica}')
```
